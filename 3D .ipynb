```python
%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
```

```python
M=1#0.51*10**6

A=500
A1=30
```

```python
#Matrices que se llenaran
X=zeros((A,A))#maximo valor para X es 1 para tomar problema de tunelamiento E/V_{0}
E=zeros((A,A))#valor de las energias w que se tomara
B=zeros((A,A))#valor de a_{w}**2/w
a=0.01


#valores maximos en X,Y y Z
z=[]
for i in range(A1):
    z.append(i*1.0/A1)
    #if(i*1.0/A+0.001>0.5):
    #z.append(i*1.0/A+0.001)
x=[]
for i in range(A):
    #if(i*1.0/A+0.001>0.5):
    x.append(i*1.0/A+0.001)
    
#print "minimos x,z",min(x),min(z),"maximos x,z",max(x),max(z),len(x),len(z)


for i in range(len(z)):
    for j in range(len(x)):
        
        if(x[j]>=z[i]):
            b=0
        if(x[j]<z[i]):
            Dt=a*sqrt(M/(2*1-2*z[i]))
            #Dt=a*sqrt(M/(2*1-2*X1+2*Egamma1[i]))
            a1=( ( (  sqrt(2*(1-z[i]+x[j])/M)   -   sqrt(2*z[i]/M)      )**2 )/x[j])
            a2=(((  sqrt(2*(z[i]-x[j])/M)   -     sqrt(2*(1-z[i])/M)  )**2 )/x[j])
            a3=( 2*(cos(x[j]*Dt)/x[j])* ((sqrt(2*(1-z[i]+x[j])/M)   -  sqrt(2*z[i]/M))*(sqrt(2*(z[i]-x[j])/M)   - sqrt(2*(1-z[i])/M))) )            
            b=a1+a2+a3
 
            
        X[i][j]=z[i]
        E[i][j]=x[j]
        B[i][j]=b #sin(2*pi*x[i])*sin(2*pi*x[j])
        
        
            
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X,E, B, rstride=10, cstride=10)
ax.set_zlim(-.4, 0.4)





plt.show()    

#print shape(X),shape(E),shape(B)
#print "X es \n",X,"\n E es \n", E
```
