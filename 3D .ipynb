```python
%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
```

```python
M=1#0.51*10**6
Dt=10
A=500
A1=50
```

```python
#Matrices que se llenaran
X=zeros((A,A))#maximo valor para X es 1 para tomar problema de tunelamiento E/V_{0}
E=zeros((A,A))#valor de las energias w que se tomara
B=zeros((A,A))#valor de a_{w}**2/w

#valores maximos en X,Y y Z
z=[]
for i in range(A1):
    z.append(i*1.0/A1)
    #if(i*1.0/A+0.001>0.5):
    #z.append(i*1.0/A+0.001)
x=[]
for i in range(A):
    #if(i*1.0/A+0.001>0.5):
    x.append(i*1.0/A+0.001)
    
#print "minimos x,z",min(x),min(z),"maximos x,z",max(x),max(z),len(x),len(z)


for i in range(len(z)):
    for j in range(len(x)):
        
        if(x[j]>=z[i]):
            b=0
        if(x[j]<z[i]):
            a1=(1/M)*((2.0/x[j]+2) -sqrt(4.0*z[i]*((1-z[i])/x[j]**2+1/x[j]))    )#-sqrt(4.0*z[i]*((1-z[i])/x[j]**2+1/x[j]))
            a2=(4.0/x[j]-4*z[i]/x[j]+2-sqrt(4.0*(1-z[i])*((1-z[i])/x[j]**2+1/x[j])))/M
            a3=(2.0/M)*cos(x[j]*Dt)*(2.0*((1-z[i])/x[j]+1)+sqrt(4*z[i]*(1-z[i])/x[j]**2)-(sqrt(1-z[i])+sqrt(z[i]))*sqrt(4*(1/x[j] + (1-z[i])/x[j]**2))  )
            b=a1+a2+a3
 
            
        X[i][j]=z[i]
        E[i][j]=x[j]
        B[i][j]=a3 #sin(2*pi*x[i])*sin(2*pi*x[j])
        
        
            
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X,E, B, rstride=10, cstride=10)






plt.show()    

#print shape(X),shape(E),shape(B)
#print "X es \n",X,"\n E es \n", E
```
