```python
%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

```

```python
M=0.51*10**6

A=100
A1=60


```

```python
#Matrices que se llenaran
X=zeros((A,A))#maximo valor para X es 1 para tomar problema de tunelamiento E/V_{0}
E=zeros((A,A))#valor de las energias w que se tomara
B=zeros((A,A))#valor de a_{w}**2/w
a=0.01


#valores maximos en X,Y y Z
z=[]
for i in range(A1):
        z.append(1.1+(i)*1.0/A1)
    #if(i*1.0/A+0.001>0.5):
    #z.append(i*1.0/A+0.001)
x=[]
for i in range(A):
    #if(i*1.0/A+0.001>0.5):
    x.append((i*2.0/A+0.001)*00.1)
    
print "minimos x,z",min(x),min(z),"maximos x,z",max(x),max(z),len(x),len(z)


for i in range(len(z)):
    for j in range(len(x)):
        
        if(x[j]>=z[i]*0.1  ):
            b=0
        if(x[j]<z[i]*0.3 ):
            Dt=a*sqrt(M/(-2*1+2*z[i]-2*x[j]))
            #Dt=a*sqrt(M/(2*1-2*X1+2*Egamma1[i]))
            a1=( ( (  sqrt(2*(-1+z[i]-x[j])/M)  -  sqrt(2*z[i]/M)      )**2 )/x[j])
            a2=(((  sqrt(2*(z[i]-x[j])/M)   -     sqrt(2*(-1+z[i])/M)  )**2 )/x[j])
            a3=abs( 2*(cos(x[j]*Dt)/x[j])* (sqrt(2*(-1+z[i]+x[j])/M)   -  sqrt(2*z[i]/M)) * (sqrt(2*(z[i]-x[j])/M)   - sqrt(2*(-1+z[i])/M)) )            
            b=a1+a2+a3
 
            
        X[i][j]=z[i]
        E[i][j]=x[j]
        B[i][j]=b #sin(2*pi*x[i])*sin(2*pi*x[j])
        
 

```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')

Y = E
Z =B
surf = ax.plot_surface(Y,X, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
        linewidth=0, antialiased=False)
#ax.set_xlim(0,0.1)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
ax.set_ylabel("$ E/V_{0}$",size=20)
ax.set_xlabel("$ E_{\gamma}/ V_{0} $",size=20)
ax.set_zlabel("$ E_{\gamma}^{-1}|a_{\omega}|^{2} $",size=20)

ax.invert_yaxis()
#ax.set_xlim(0,0.005)

fig.colorbar(surf, shrink=0.5, aspect=5)
figsize(20,8)

#savefig("3d_despues.png")
plt.show()





```
