```python
X1=0.5 # Ee/Vo
M=1#0.51*10**6
Dt=1000000
Bremss=[]
```

```python
Egamma1=[]

for i in range(1000):
    if(i*0.001+0.4<=X1): #cota maxima de la energia de Egamma es X1 
        Egamma1.append(i*0.001+0.0001)
print max(Egamma1)   
```

# $ funciones\ $

```python
a1=[]
a2=[]
a3=[]
A=[]

for i in range(len(Egamma1)):
    a1.append(( sqrt(2*(1-X1+Egamma1[1])/M) - sqrt(2*(X1)/M))**2/Egamma1[i])
    a2.append(( sqrt(2*(1-X1+Egamma1[1])/M) - sqrt(2*(1-X1)/M))**2/Egamma1[i])
    a3.append((sqrt(2*(1-X1+Egamma1[1])/M) - sqrt(2*(1-X1)/M))*( sqrt(2*(1-X1+Egamma1[1])/M) - sqrt(2*(X1)/M))*cos(Egamma1[i]*Dt)/Egamma1[i])
    A.append(a1[i]+a2[i]+a3[i])
print len(A),len(a1),len(a2),len(a3),len(Egamma1)
```

```python
sqrt(2*(1-X1+Egamma1[1])/M)
plot(Egamma1,a1,"r",label="$ first\  term\ $")
plot(Egamma1,a2,"y--",label="$ second\ term\ $")
plot(Egamma1,a3,"b-",label="$ Oscilatory\ term\ $")
plot(Egamma1,A,"k-",label="$|a_{\omega}|^{2}(E_{\gamma}/V_{0})^{-1}$")
xlabel("$ E_{\gamma}/V_{0} $",size=20)
ylabel("$ f(E_{\gamma}/V_{0}) $",size=20)
legend()
ylim(min(a3)-0.0001,.0005)
```
