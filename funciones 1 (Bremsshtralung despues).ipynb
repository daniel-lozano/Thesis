```python
a=0.01

for k in range(10):
    X1=0.1*k+0.09
    
    a1=[]
    a2=[]
    a3=[]
    A=[]
    Egamma1=[]

    for i in range(10000):
        if(i*0.001<X1): #cota maxima de la energia de Egamma es X1 
            Egamma1.append(i*0.001+0.0001)
     
    for i in range(len(Egamma1)):
        Dt=a*sqrt(M/(2*1-2*X1))
        a1.append( ( (  sqrt(2*(1-X1+Egamma1[i])/M)   -   sqrt(2*X1/M)      )**2 )/Egamma1[i])
        a2.append( ((  sqrt(2*(X1-Egamma1[i])/M)   -     sqrt(2*(1-X1)/M)  )**2 )/Egamma1[i])
        a3.append( abs(2*(sqrt(2*(1-X1+Egamma1[i])/M)   -  sqrt(2*X1/M))*(sqrt(2*(X1-Egamma1[i])/M)   - sqrt(2*(1-X1)/M)))*cos(Egamma1[i]*Dt)/Egamma1[i]  )
        A.append(a1[i]+a3[i]+a3[i])
    

        #print len(A),len(a1),len(a2),len(a3),len(Egamma1)
    plt.show()
    plot(Egamma1,a1,"r",label="$ first\  term\ $")
    plot(Egamma1,a2,"m",label="$ second\ term\ $")
    plot(Egamma1,a3,"b-",label="$ Oscilatory\ term\ $")
    plot(Egamma1,A,"k-",label="$|a_{\omega}|^{2}(E_{\gamma}/V_{0})^{-1}$")
    xlabel("$ E_{\gamma}/V_{0} $",size=20)
    ylabel("$ f(E_{\gamma}/V_{0}) $",size=20)
    title("$ E_{\gamma}/V_{0}= $"+str(X1))
    ylim(-0.001,0.001)#ylim(-2,2)
    legend(loc=4)
    
ylim(-0.0001,0.0001)
legend(loc=4)

```
