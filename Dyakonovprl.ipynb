```python
M=0.51*10**6
Dt=300
print abs(-1)
```

# $ Bremsshtralung\ after\ $

```python
a=0.01

for k in range(3):
    if(k==0):
        X1=0.1
    if(k==1):
        X1=0.5
    if(k==2):
        X1=0.9
    
    a1=[]
    Egamma1=[]

    for i in range(10000):
        if(i*0.001*0.1<X1*0.1): #cota maxima de la energia de Egamma es X1 
            Egamma1.append(i*0.001*0.1+0.0001)
     
    for i in range(len(Egamma1)):
        Dt=a*sqrt(M/(2*1-2*X1))
        a1.append( (Egamma1[i]*Dt)**2 )
      
    

        #print len(A),len(a1),len(a2),len(a3),len(Egamma1)
    plt.show()
    plot(Egamma1,a1,"r",label="$ Dyakonov\ for\ high\ \omega $")
    xlabel("$ E_{\gamma}/V_{0} $",size=20)
    ylabel("$ f(E_{\gamma}/V_{0}) $",size=20)
    title("$ E_{\gamma}/V_{0}= $"+str(X1))
    #ylim(-0.001,0.001)#ylim(-2,2)
    legend(loc=4)
    
legend(loc=4)


```

# $ Bremsshtralung\ Before\ $

```python
a=0.01

for k in range(3):
    if(k==0):
        X1=0.1
    if(k==1):
        X1=0.5
    if(k==2):
        X1=0.9
    
    a1=[]
    Egamma1=[]

    for i in range(10000):
        if(i*0.001*0.1<X1*0.1): #cota maxima de la energia de Egamma es X1 
            Egamma1.append(i*0.001*0.1+0.00001)
     
    for i in range(len(Egamma1)):
        Dt=a*sqrt(M/(2*1-2*X1+2*Egamma1[i]))
        a1.append( (Egamma1[i]*Dt)**2 )
      
    

        #print len(A),len(a1),len(a2),len(a3),len(Egamma1)
    plt.show()
    plot(Egamma1,a1,"r",label="$ Dyakonov\ for\ high\ \omega $")
    xlabel("$ E_{\gamma}/V_{0} $",size=20)
    ylabel("$ f(E_{\gamma}/V_{0}) $",size=20)
    title("$ E_{\gamma}/V_{0}= $"+str(X1))
    #ylim(-0.001,0.001)#ylim(-2,2)
    legend(loc=4)
    
legend(loc=4)



```
